## Investigation Overview


> In this exploration, I wanted to understand what is most associated with original loan amount. I focused on few variables I thought would be decisive such as monthly loan payment, and service fee.



## Dataset Overview

> The prosper loan dataset consists  113,936 debt data distributed across 81 features. since the focus features don't have missing values, none was dropped. however, the lessen the effect of extreme outliers few transformation has be performed.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan =pd.read_csv("prosperLoanData.csv")
```

## (Visualization 1)
 

## The distribution of original loan amount
The loan amount have wide distribution strecthing from the lowest 1000 to highest 35,000. plotted on log scale, loan amount distribution is slightly left skewed with sharp raise around 10,000 and 15,000 - 20,000.

```python
# put on log scale to eliminate long and fragmented tail of original loan amount feature

bins = 10**np.arange(3, np.log10(loan["LoanOriginalAmount"].max())+0.12, 0.12)

plt.figure(figsize=[10, 6])
plt.hist(data=loan, x='LoanOriginalAmount', bins = bins)
plt.xscale('log')
plt.xticks([500, 1e3, 5e3, 1e4, 15e3, 2e4, 25e3, 3e4, 35e3, 4e4], [500, '1k', '10k' ,'15k','20k','25k','30k','35k','40k'])
plt.xlabel('Original amount of loan [$]')
plt.title('The distribution of the original amount of loans [$]')
plt.show();
```

## (Visualization 2)

## The distribution of monthly loan payment
Monthly loan payment stretchs from 0 upto 2251. and the extreme outliers were trimmed using xlim. the resulting distribution was slightly right skewed with most of its values concentrated on lower end. 

```python
# plot histogram to see the distribution of mothly payments
bins = np.arange(0, loan["MonthlyLoanPayment"].max()+50, 50)

plt.hist(data=loan, x='MonthlyLoanPayment', bins = bins)
plt.xlabel('monthly payment [$]')
plt.xlim([0, 1000])

plt.title('The distribution of montly payment');
```

## (Visualization 3)

## Original loan amount Vs monthly loan payment
ploting original loan amount against monthly loan payment in log scale showed linear relationship between them. furthermore, they have strong positive correlation with each other. 

```python
# scatter plot of price vs. monthly loan payment, with log transform of both axis by log scale.

plt.figure(figsize = [12, 8])
plt.scatter(data = loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', alpha = 1/20)

plt.xscale('log')
plt.xticks([10,50,100,250,500,750, 1e3,], [10, 50, '0.1k', '0.5k', '0.75k','1k'])
plt.xlabel('monthly payment ($)')

plt.yscale('log')
plt.yticks([500, 1e3, 2e3, 5e3, 1e4, 2e4], [500, '1k', '2k', '5k', '10k', '20k'])
plt.ylabel('amount of loan ($)')

plt.title('Original loan amount Vs Monthly loan Payment')

plt.show();
```

## (Visualization 4)

## original loan amount by monthly payment and service fee
when plotted on log scale, original loan amount positively correlated with monthly payment where as service fee is negatively correlated with both, but not strongly.

```python
# ploting scatter with color for third variable
plt.figure(figsize = [12, 8])

plt.scatter(data = loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', c = 'LP_ServiceFees')
plt.colorbar()

# transforming to log scales 
plt.xscale('log')
plt.xticks([10,50,100,250,500,750, 1e3,], [10, 50, '0.1k', '0.5k', '0.75k','1k'])
plt.xlabel('monthly payment ($)')

plt.yscale('log')
plt.yticks([500, 1e3, 2e3, 5e3, 1e4, 2e4], [500, '1k', '2k', '5k', '10k', '20k'])
plt.ylabel('amount of loan ($)')

plt.title ('original loan amount Vs monthly loan payment Vs service fee')

plt.show();
```
